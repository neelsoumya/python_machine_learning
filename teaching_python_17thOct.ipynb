# Do PCA

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

# Load new data

```python
url = "https://raw.githubusercontent.com/cambiotraining/ml-unsupervised/main/course_files/data/USArrests.csv"
X = pd.read_csv(url, index_col=0)

print(X.head())
```

# Normalize the data

```python
scaler_standard = StandardScaler()
X_scaled = scaler_standard.fit_transform(X)

print(X_scaled)
```

# Perform PCA

```python
pca_fn = PCA()
X_pca = pca_fn.fit_transform(X_scaled)

print(X_pca)
```

# Plot biplot

```python
plt.figure()
plt.scatter(X_pca[:,0], X_pca[:,1])
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("PCA Biplot")
plt.show()
```

# Categorical variables

```python
print(X.index)

states = X.index

state_codes = pd.Categorical(states).codes

plt.figure()
plt.scatter(X_pca[:,0], X_pca[:,1], c=state_codes)
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("PCA biplot")
plt.show()

```

# getting labels

```python
plt.figure()
plt.scatter(X_pca[:,0], X_pca[:,1], alpha = 0.7)
plt.show()

plt.figure()
plt.scatter(X_pca[:,0,], X_pca[:,1], alpha = 0.7)

print(X.index)
```

# definitive code to plot text

```python
fig, ax = plt.subplots()

ax.scatter(X_pca[:,0], X_pca[:,1])

for i, state in enumerate(X.index):
  ax.text(X_pca[i,0], X_pca[i,1], state, fontsize = 6)

ax.set_xlabel("PC1")
ax.set_ylabel("PC2")
ax.set_title("PCA biplot with text")

plt.show()
```

# getting the loadings

```python
print(pca_fn.components_.T)

loadings = pca_fn.components_.T

pd.DataFrame(loadings,
             index = X.columns
             )
```

# Using the PCA package

```python
from pca import pca

url = "https://raw.githubusercontent.com/cambiotraining/ml-unsupervised/main/course_files/data/USArrests.csv"

df = pd.read_csv(url, index_col=0)

print(df.head())
```

# Normalize and perform PCA

```python
model = pca(normalize = True)

out = model.fit_transform(df)

ax = model.biplot()
```

# loadings

```python
print(model.results)
```

```python
# Extract the desired keys and convert them to DataFrames
loadings_df = pd.DataFrame(model.results['loadings'])
pc_df = pd.DataFrame(model.results['PC'])
explained_var_df = pd.DataFrame({'explained_variance': model.results['explained_var']})
variance_ratio_df = pd.DataFrame({'variance_ratio': model.results['variance_ratio']})
topfeat_df = pd.DataFrame(model.results['topfeat'])
outliers_df = pd.DataFrame(model.results['outliers'])

# Display the DataFrames
print("Loadings DataFrame:")
display(loadings_df)

print("\nPrincipal Components DataFrame:")
display(pc_df)

print("\nExplained Variance DataFrame:")
display(explained_var_df)

print("\nVariance Ratio DataFrame:")
display(variance_ratio_df)

print("\nTop Features DataFrame:")
display(topfeat_df)

print("\nOutliers DataFrame:")
display(outliers_df)
```

```python
model.plot()
```

# Load the data

```python
from sklearn import datasets
import umap
import seaborn as sns
import pandas as pd
```

```python
sns.set(style="whitegrid")

# load the dataset
iris = datasets.load_iris()

X_iris = iris.data
y_iris = iris.target

labels_iris = iris.target_names

# create a data frame and store these vales
df_iris_simple = pd.DataFrame(X_iris, columns = iris.feature_names)

df_iris_simple['species'] = iris.target
df_iris_simple['species_name'] = df_iris_simple['species'].map( {0:'setosa', 1:'versicolor', 2:'virginica'} )

print(df_iris_simple.head)
```

# Visualize this data

```python
plt.figure()
plt.scatter( X_iris[:,0], X_iris[:,1], c = iris.target, cmap = 'viridis' )
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.show()
```

# Normalize your data

```python
scaler = StandardScaler()
X_iris_scaled = scaler.fit_transform(X_iris)
```

# Apply UMAP

```python
umap_model = umap.UMAP(n_neighbors = 15, min_dist=0.1, n_components=2)
embedding_iris = umap_model.fit_transform(X_iris_scaled)

```

```python
df_iris_components = pd.DataFrame({
    "UMAP1": embedding_iris[:,0],
    "UMAP2": embedding_iris[:,1],
    "species": [ labels_iris[i] for i in y_iris ]
})

plt.figure()
sns.scatterplot(df_iris_components, x="UMAP1", y="UMAP2", hue="species")
plt.title("UMAP plot")
plt.show()
```

# k-means

```python
from sklearn.cluster import KMeans
```

```python
print(df_iris_simple.head)
```

```python
# load the dataset
iris = datasets.load_iris()

X_iris = iris.data
y_iris = iris.target
```

## Perform k-means

```python
kmeans = KMeans(n_clusters=2, n_init = 20)
kmeans.fit(X_iris)
```

# Plot

```python
plt.figure()
plt.scatter(X_iris[:,0], X_iris[:,1], c=kmeans.labels_)
plt.title("K-means on Iris data")
plt.show()
```

# Pick your hyperparameter $K$

```python
wcss = []

# try different values of K
k_range = range(1,11)

for k_var in k_range:

  kmeans = KMeans(n_clusters = k_var)
  kmeans.fit(X_iris)

  wcss.append(kmeans.inertia_)
```

# Plot

```python
plt.figure()
plt.scatter(k_range, wcss)
plt.xlabel('Values of $K$')
plt.ylabel('WCSS')
plt.show()
```

# Hierarchical clustering

```python
from scipy.cluster.hierarchy import linkage, dendrogram
from sklearn.cluster import AgglomerativeClustering
```

# fit a dendrogram

```python
agg = AgglomerativeClustering(n_clusters=3, linkage="ward")
labels = agg.fit(X_iris)
```

```python
Z = linkage(X_iris, method="ward")
```
