## Teaching hierarchical clustering in Python

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from scipy.cluster.hierarchy import linkage, dendrogram
from sklearn.cluster import AgglomerativeClustering
```

```python
# Load the Iris dataset (biological example)
iris = load_iris()
X = iris.data
df = pd.DataFrame(X, columns=iris.feature_names)

# 1. Compute the linkage matrix using Ward's method
Z = linkage(X, method='ward')

# 2. Plot the dendrogram
plt.figure(figsize=(10, 5))
dendrogram(Z, labels=iris.target)
plt.title('Hierarchical Clustering Dendrogram (Iris)')
plt.xlabel('Sample Index or (Cluster Size)')
plt.ylabel('Distance')
plt.show()

# 3. Perform Agglomerative Clustering with 3 clusters
hc = AgglomerativeClustering(n_clusters=3, linkage='ward')
labels = hc.fit_predict(X)
df['cluster'] = labels

# 4. Display the first few rows with assigned cluster labels
print(df.head())
```
